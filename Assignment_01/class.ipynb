```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"/content/MLUP-class/Assignment_01/auto-mpg.csv"
```

### Load Auto MPG datasets

```python
def load_auto_mpg(file_path):
    # column_names = [
    #     "mpg", "cylinders", "displacement", "horsepower",
    #     "weight", "acceleration", "model_year", "origin", "car_name"
    # ]

    try:
        df = pd.read_csv(
            file_path,
            # sep=r"\s+",
            # names=column_names,
            na_values='?',
        )

        print("\nDataset loaded successfully!")
        print(f"Shape: {df.shape}")
        print("\nFirst 5 rows:")
        print(df.head())

        missing_count = df.isnull().sum()
        if missing_count.any():
            print("\nMissing values detected:")
            print(missing_count[missing_count > 0])
        else:
            print("\nNo missing values found.")

        return df

    except Exception as e:
        print("Error loading dataset:", e)
        return None

df = load_auto_mpg(file_path)
```

### Detect Outliers in HorsePower

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

Q1= df['horsepower'].quantile(0.25)
Q3= df['horsepower'].quantile(0.75)
IQR= Q3-Q1

lower= Q1-1.5*IQR
upper= Q3+1.5*IQR

min_hp= df['horsepower'].min()
max_hp= df['horsepower'].max()
print("Minimum horsepower:", min_hp)
print("Maximum horsepower:", max_hp)

outliers= df[(df['horsepower']<lower) | (df['horsepower']>upper)]
print("Number of Outliers: ", len(outliers))
```

### Remove Outliers

```python
df_clean= df[(df['horsepower']>=lower) & (df['horsepower']<=upper)]
```

## Handling Missing values

### Remove Records
Good when missing values are very few

```python
print("Before removing the records: ", len(df))
df_drop= df.dropna()
print("Remaining Records: ", len(df_drop))
```

### Mean Imputation
Replace missing horsepower with mean value

```python
mean_hp= df['horsepower'].mean()
df['horsepower'].fillna(mean_hp, inplace=True)
print(df['horsepower'])
```

### Median Imputation
Better when outliers exist

```python
median_hp= df['horsepower'].median()
df['horsepower'].fillna(median_hp, inplace= True)
```
