### Time taken to load Iris Dataset

```python
iris=datasets.load_iris()
iris_df= pd.DataFrame(iris.data,columns= iris.feature_names)
iris_df['target']=iris.target
print(pd.concat([iris_df.head(3),iris_df.tail(3)]))
```

```python
import sklearn.model_selection as skms

X_train, X_test, y_train, y_test = skms.train_test_split(
    iris.data,iris.target,test_size=0.2)

print("Train features: ", X_train.shape)
print("Train labels: ", y_train.shape)
print("Test features shape: ", X_test.shape)
```

```python
from sklearn import naive_bayes
from sklearn import neighbors, metrics
```

```python
def run_nb():
  nb= naive_bayes.GaussianNB()
  fit=nb.fit(X_train,y_train)
  preds=fit.predict(X_test)
```

```python
def run_knn():
  knn= neighbors.KNeighborsClassifier(n_neighbors=3)
  fit= knn.fit(X_train,y_train)
  preds= fit.predict(X_test)
```

```python
nb_time1= timeit.repeat(run_nb,repeat=1,number=1)
knn_time1= timeit.repeat(run_knn,repeat=1,number=1)
print("NB 1 time: ",nb_time1)
print("KNN 1 time: ",knn_time1)
```

```python
nb_time1000= timeit.repeat(run_nb,repeat=1,number=1000)
knn_time1000= timeit.repeat(run_knn,repeat=1,number=1000)
print("NB 1000 time: ",nb_time1000)
print("KNN 1000 time: ",knn_time1000)
```
